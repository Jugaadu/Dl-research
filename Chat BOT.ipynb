# Data Preperation

```python
# load packages
import os
import json
import nltk
import gensim
import numpy as np
from gensim import corpora, models, similarities
import pickle
```

```python
model = gensim.models.Word2Vec.load('/home/jb/Documents/chatbot/apnews_sg/word2vec.bin')
```

```python
file=open('/home/jb/Documents/chatbot/corpus/conversations.json')
```

```python
data = json.load(file)
```

```python
cor=data["conversations"]
```

```python
x=[]
y=[]


for i in range(len(cor)):
    for j in range(len(cor[i])):
        if j<len(cor[i])-1:
            x.append(cor[i][j]);
            y.append(cor[i][j+1]);
```

```python
tok_x=[]
tok_y=[]
for i in range(len(x)):
    tok_x.append(nltk.word_tokenize(x[i].lower()))
    tok_y.append(nltk.word_tokenize(y[i].lower()))
```

```python
sentend=np.ones((300,),dtype=np.float32)
```

```python
vec_x=[]
for sent in tok_x:
    sentvec = [model[w] for w in sent if w in model.vocab]
    vec_x.append(sentvec)
```

```python
vec_y=[]
for sent in tok_y:
    sentvec = [model[w] for w in sent if w in model.vocab]
    vec_y.append(sentvec)
```

```python
for tok_sent in vec_x:
    tok_sent[14:]=[]
    tok_sent.append(sentend)
```

```python
for tok_sent in vec_x:
    if len(tok_sent)<15:
        for i in range(15-len(tok_sent)):
            tok_sent.append(sentend)
```

```python
for tok_sent in vec_y:
    tok_sent[14:]=[]
    tok_sent.append(sentend)
```

```python
for tok_sent in vec_y:
    if len(tok_sent)<15:
        for i in range(15-len(tok_sent)):
            tok_sent.append(sentend)
```

```python
with open('conversation.pickle','wb') as f:
    pickle.dump([vec_x,vec_y],f)
```

```python
model=load_model('LSTM5000.h5')
mod = gensim.models.Word2Vec.load('/home/jb/Documents/chatbot/apnews_sg/word2vec.bin')
```

```python
while(True):
    x=input("You: ")
    sentend=np.ones((300,),dtype=np.float32) 

    sent=nltk.word_tokenize(x.lower())
    sentvec = [mod[w] for w in sent if w in mod.vocab]

    sentvec[14:]=[]
    sentvec.append(sentend)
    if len(sentvec)<15:
        for i in range(15-len(sentvec)):
            sentvec.append(sentend) 
    sentvec=np.array([sentvec])
    
    predictions = model.predict(sentvec)
    outputlist=[mod.most_similar([predictions[0][i]])[0][0] for i in range(15)]
    output=' '.join(outputlist)
    output = output.replace('karluah','').replace('kleiser','')
    print("Bot:", output)
```
